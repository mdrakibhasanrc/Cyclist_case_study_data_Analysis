# Data Cleaning:

```python
# show columns name
df.columns
```

```python
df.head(2)
```

```python
# drop unwanted columns
df.drop(columns=['end_station_name','end_station_id'],axis=1,inplace=True)
```

```python
# change data type
df['started_at']=pd.to_datetime(df['started_at'])
df['ended_at']=pd.to_datetime(df['ended_at'])
```

```python
df[['Start_Hour','Start_Minutes','Start_Seconds']]=df['started_at'].astype('str').str.split(':',expand=True)
```

```python
df['Start_hour']=df['Start_Hour'].astype('str').str.split(' ').str[1]
```

```python
df['Start_Date']=df['Start_Hour'].astype('str').str.split(' ').str[0]
```

```python
df.drop(columns=['started_at','ended_at','end_lat','end_lng','Start_Hour'],axis=1,inplace=True)
```

```python
df.drop(columns=['Start_Minutes','Start_Seconds'],axis=1,inplace=True)
```

```python
df['Start_Date']=pd.to_datetime(df['Start_Date'])
```

```python
df['Day_of_week']=df['Start_Date'].dt.day_name()
```

```python
df['Month']=df['Start_Date'].dt.month_name()
```
